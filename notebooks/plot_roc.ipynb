```python
import sys
import os

# Get the absolute path to the src directory
src_path = os.path.abspath(os.path.join('..', 'src'))
if src_path not in sys.path:
    sys.path.append(src_path)

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utils.figure_style import FigureStyle
from utils.json_data_reader import JSONDataReader
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize

# Access constants from JSON
json_reader = JSONDataReader('../data/config.json')
sample_names = json_reader.get('sample_names')
group_names = json_reader.get('group_names')
sample_colors = json_reader.get('sample_colors')
group_colors = json_reader.get('group_colors')
fig_dims = json_reader.get('figure_dimensions')
one_third_fig_width = fig_dims.get('one_third_fig_width')
half_fig_width = fig_dims.get('half_fig_width')
full_fig_width = fig_dims.get('full_fig_width')
fig_height_a0 = fig_dims.get('fig_height_a0')
fig_height_a = fig_dims.get('fig_height_a')
fig_height_b = fig_dims.get('fig_height_b')
```

```python
np.random.seed(2)
n_samples = 120
df = pd.DataFrame({
    'Feature1': np.random.randn(n_samples),
    'Feature2': np.random.randn(n_samples),
    'Group': np.random.choice(group_names, n_samples)
})

# Binarize group labels
y_true = label_binarize(df['Group'], classes=group_names)[:, 0]
y_scores = np.random.rand(n_samples)

# Compute ROC curve and ROC area
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

# Create an instance of FigureStyle with custom colors
style_config = FigureStyle(colors=group_colors)
style_config.apply()

# Plot ROC curve
fig, ax = plt.subplots(figsize=(half_fig_width*0.9, fig_height_a))
ax.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.2f})', color=group_colors[group_names[1]])
ax.plot([0, 1], [0, 1], linestyle='--', color='grey')

# Customize the plot
style_config.set_titles(ax, 'ROC Curve 2')
style_config.set_labels(ax, 'False Positive Rate', 'True Positive Rate')
style_config.update_legend_style(ax, loc='lower right')

# Save the figure
style_config.save_figure('../figures/roc_curve1.pdf')

# Show the plot (optional)
plt.show()
```
